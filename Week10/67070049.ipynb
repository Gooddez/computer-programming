```python
# ex1
class Employee:
    name = 'name'
    age = 0
    position = 'position'
    def display_info(self):
        print(f'Name: {self.name}')
        print(f'Age: {self.age}')
        print(f'Position: {self.position}')
```

```python
#ex2
class FullTimeEmployee(Employee):
    monthly_salary = 0
    bonus_percentage = 0
    performance_rating = 0
    def calculate_annual_salary_with_bonus(self):
        bonus = self.monthly_salary * self.bonus_percentage/100 * self.performance_rating/5 
        annual_salary = self.monthly_salary + bonus
        return annual_salary
    def apply_raise(self,raise_percentage):
        new_salary = (raise_percentage/100*self.monthly_salary) + self.monthly_salary
        self.monthly_salary = new_salary
    def compare_performance(self,other_employee):
        if isinstance(other_employee,FullTimeEmployee):
            if self.performance_rating > other_employee.performance_rating:
                return f'{self.name} has better performance than {other_employee.name}.'
            elif self.performance_rating < other_employee.performance_rating:
                return f'{other_employee.name} has better performance than {self.name}.'
            else:
                return 'Both performance are equal.'

alice = FullTimeEmployee()
alice.name = 'Alice'
alice.age = 30
alice.position = 'Software Engineer'
alice.monthly_salary = 50000
alice.bonus_percentage = 10
alice.performance_rating = 4

bob = FullTimeEmployee()
bob.name = 'Bob'
bob.age = 35
bob.position = 'Team Lead'
bob.monthly_salary = 60000
bob.bonus_percentage = 15
bob.performance_rating = 5

print('Full-Time Employee Information:')
alice.display_info()
print(f'Annual Salary with Bonus: {alice.calculate_annual_salary_with_bonus()}')
alice.apply_raise(5)
print(f'Monthly Salary after 5% raise: {alice.monthly_salary}')
print(alice.compare_performance(bob))
```

```python
# ex3
class PartTimeEmployee(Employee):
    hourly_rate = 0
    hours_worked = 0
    overtime_hours = 0
    def calculate_monthly_income_with_overtime(self):
        return (self.hourly_rate * self.hours_worked) + (self.overtime_hours*self.hourly_rate*1.5)
    def add_hours(self,additional_hour):
        self.hours_worked += additional_hour
    def compare_income(self,other_employee):
        if isinstance(other_employee,PartTimeEmployee):
            if self.calculate_monthly_income_with_overtime() > other_employee.calculate_monthly_income_with_overtime():
                return f'{self.name} earn more that {other_employee.name}.'
            elif self.calculate_monthly_income_with_overtime() < other_employee.calculate_monthly_income_with_overtime():
                return f'{other_employee.name} earn more that {self.name}.'
            else:
                return 'Both performance are equal.'
ton = PartTimeEmployee()
ton.name = 'ton'
ton.age = 30
ton.position = 'Software Engineer'
ton.hourly_rate = 200
ton.hours_worked = 100
ton.overtime_hours = 10

tri = PartTimeEmployee()
tri.name = 'tri'
tri.age = 35
tri.position = 'Team Lead'
tri.hourly_rate = 250
tri.hours_worked = 90
tri.overtime_hours = 15

print('Part-Time Employee Information:')
ton.display_info()
print(f'monthly income with overtime: {ton.calculate_monthly_income_with_overtime()}')
ton.add_hours(20)
print(f'hours worked after adding 20 hours: {ton.hours_worked}')
print(ton.compare_income(tri))
```

```python
class book:
    title = 'title'
    isbn = 'xxx-xx-xxxxx-xx-x'
    author = 'author'
    publisher = 'publisher'
    price = 0
    def sample(self):
        if isinstance(self,paperback) or isinstance(self,ebook):
            samplepage = random.randint(1,self.print_length - 9)
            for page in range(samplepage,samplepage+10):
                print(page)
        if isinstance(self,audiobook):
            minute = int(self.listening_length[:2])
            second = int(self.listening_length[3:])
            all = minute*60+second
            start = random.randint(0,all - 9)
            end = start+10
            print(f'{start//60:>02}:{start%60:>02}-{end//60:>02}:{end%60:>02}')
            
    def calshipping(self):
        if isinstance(self,paperback):
            if self.dimension >= 1000:
                print(self.weight*0.25+20)
            else:
                print(self.weight*0.1)
        if isinstance(self,ebook) or isinstance(self,audiobook):
            if self.file_size >= 100:
                print(self.file_size*0.25+20)
            elif self.file_size < 100:
                print(self.file_size*0.1)

# ex4
class paperback(book):
    print_length = 0
    dimension = 0
    weight = 0

# ex5
class ebook(book):
    file_size = 0
    print_length = 0
    current_page = 0
    def next(self):
        self.current_page += 1
        print(self.current_page)
    def previous(self):
        self.current_page -= 1
        print(self.current_page)

# ex6
class audiobook(book):
    listening_length = '00:00'
    narrator = 'narrator'
    file_size = 0
    current_time = '00:00'
    
    def play(self):
        all = (int(self.listening_length[:2])*60)+int(self.listening_length[3:])
        now = (int(self.current_time[:2])*60)+int(self.current_time[3:])
        count = 0
        while now <= all and count <= 15:
            print(f'{now//60:>02}:{now%60:>02}')
            now += 1
            count += 1

```
