```python
# exercises 1
class Triangle:
    def __init__(self, length):
        self.length = length
        self.area = (math.sqrt(3) / 4)*(self.length**2)
        self.perimeter = self.length * 3
    def calArea(self):
        return self.area
    def calPerimeter(self):
        return self.perimeter
    def __str__(self):
        obj = 'Object: Equilaterala triangle\n'
        side = f'Side length: {self.length:.2f} cm\n'
        area = f'Area: {self.area:.2f} cm\u00B2\n'
        perimeter = f'Perimeter: {self.perimeter:.2f} cm'
        return obj + side + area + perimeter
    def __lt__(self, triangle1):
        return self.area < triangle1.area
```

```python
# exercises 1 TEST
obj_1 = Triangle(2)
obj_2 = Triangle(3)
print(obj_1)
print('-------------------------')
print(obj_2)
print('-------------------------')
print(obj_1 < obj_2)
```

```python
# exercises 2
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.area = self.width * self.height
        self.perimeter = (self.width * 2) + (self.height * 2)
    def calArea(self):
        return self.area
    def calPerimeter(self):
        return self.perimeter
    def __str__(self):
        obj = 'Object: Rectangle\n'
        width = f'Width: {self.width:.2f} cm\n'
        height = f'height: {self.height:.2f} cm\n'
        area = f'Area: {self.area:.2f} cm\u00B2\n'
        perimeter = f'Perimeter: {self.perimeter:.2f} cm'
        return obj + width + height + area + perimeter
    def __lt__(self, rectangle1):
        return self.area < rectangle1.area
```

```python
# exercises 2 TEST
rec_1 = Rectangle(8.2, 3.5)
rec_2 = Rectangle(6.34, 2.87)
print(rec_1)
print('-------------------------')
print(rec_2)
print('-------------------------')
print(rec_1 < rec_2)
```

```python
# exercises 3
class Circle:
    def __init__(self, radius):
        self.radius = radius
        self.diameter = radius * 2
        self.area = self.radius ** 2 * math.pi
        self.perimeter = self.radius * 2 * math.pi
    def calDiameter(self):
        return self.diameter
    def calArea(self):
        return self.area
    def calPerimeter(self):
        return self.perimeter
    def __str__(self):
        obj = f'Object: Circle\n'
        radius = f'Radius: {self.radius:.2f} cm\n'
        diameter = f'Diameter: {self.diameter:.2f} cm\n'
        area = f'Area: {self.area:.2f} cm\u00B2\n'
        perimeter =  f'Perimeter: {self.perimeter:.2f} cm'
        return obj + radius + diameter + area + perimeter
    def __lt__(self, circle1):
        return self.area < circle1.area
    
```

```python
# exercises 3 TEST
cir_1 = Circle(0.45)
cir_2 = Circle(2.64)
print(cir_1)
print('------------------------')
print(cir_2)
print('------------------------')
print(cir_1 < cir_2)
```

```python
# exercises 4
class Television:
    subscription = {
        3 : 'Channel 3',
        5 : 'Channel 5',
        7: 'Channel 7',
        9: 'Channel 9'
    }
    def __init__(self, model, subscription=None):
        self.model = model
        if subscription:
            for channel,name in enumerate(sorted(list(set(subscription)))):
                self.subscription[channel+10] = name
        self.status = False
        self.volume = 50
        self.currentChannel = 3
    def addSubscription(self, subscription):
        current = list(self.subscription.keys())[-1]
        for channel,name in enumerate(sorted(subscription)):
            if name not in self.subscription.values():
                self.subscription[channel + 1 + current] = name
    def __str__(self):
        model = f'Model: {self.model}\n'
        subscription = f'Subscription: {self.subscription}\n'
        status = f'Status: {'on' if self.status else 'off'}\n'
        volume = f'Volume: {self.volume}\n'
        channel = f'Current channel: ({self.currentChannel}) {self.subscription[self.currentChannel]}\n'
        end = '----------------------------------'
        if self.status:
            return model + subscription + status + volume + channel + end
        else:
            return model + subscription + status + end
    def turnOn(self):
        if not self.status:
            self.status = True
            print('Status: ON')
            print('----------------------------------')
        else:
            print('The TV is already turn on')
            print('----------------------------------')
    def turnOff(self):
        if self.status:
            self.status = False
            print('Status: OFF')
            print('----------------------------------')
        else:
            print('The TV is already turn off')
            print('----------------------------------')
    def changeChannel(self,newChannel):
        if self.status:
            if newChannel in self.subscription.keys():
                self.currentChannel = newChannel
                print('Change channel successful')
                print(f'Current channel: ({self.currentChannel}) {self.subscription[self.currentChannel]}')
                print('----------------------------------')
            else:
                print('Change channel unsuccessful: Unavailable channel')
                print(f'Current channel: ({self.currentChannel}) {self.subscription[self.currentChannel]}')
                print('----------------------------------')
        else:
            print('Change channel unsuccessful: The TV is turn off')
            print('----------------------------------')
    def volumeUP(self, units):
        self.volume = 100 if self.volume + units >= 100 else self.volume + units
        print(f'Volume: {self.volume}')
        print('----------------------------------')
    def volumeDOWN(self, units):
        self.volume = 0 if self.volume - units <= 0 else self.volume - units
        print(f'Volume: {self.volume}')
        print('----------------------------------')
```

```python
# exercises 5
class Mobile:
    def __init__(self,model,resolution=(int,int),battery=int,dataPlan=(int,int)):
        self.model = model
        self.resolution = resolution
        self.battery = battery
        self.dataPlan = dataPlan
        self.remainingBattery = battery
        self.minUsed = 0
        self.monthlyPayment = self.calPayment()
    def __str__(self):
        model = f'Model: {self.model}'
        resolution = f'Resolution {self.resolution[0]} x {self.resolution[1]}\n'
        battery = f'Battery: {self.battery} mAh\n'
        remaining = f'battery remaining {self.remainingBattery} ({self.remainingBattery/self.battery*100:.1f}%)\n'
        dataPlan = f'Data plan: {self.dataPlan[0]} baht/{self.dataPlan[1]} minutes\n'
        minUsed = f'Minute-used {self.minUsed} minutes\n'
        monthlyPayment = f'Monthly payment: {self.monthlyPayment} baht\n'
        end = '------------------------------------'
        return model + resolution + battery + remaining + dataPlan + minUsed + monthlyPayment + end
    def batterConsumption(self,mins):
        pixel = self.resolution[0] * self.resolution[1]
        cost = 1 if pixel<=1000000 else 2 if pixel<=1500000 else 5
        if cost * mins >= self.remainingBattery:
            able = self.remainingBattery / cost
            self.remainingBattery = 0
            return able
        else:
            self.remainingBattery -= cost*mins
            return mins
    def calPayment(self):
        pay = self.dataPlan[0] if self.minUsed <= self.dataPlan[1] else self.dataPlan[0]+((self.minUsed-self.dataPlan[1])*0.5)
        return pay
    def call(self, mins):
        if self.remainingBattery:
            able = self.batterConsumption(mins)
            self.minUsed += able
            print(f'Making a call for {able} minutes')
            print(f'battery remaining {self.remainingBattery} ({self.remainingBattery/self.battery*100:.1f}%)')
            print(f'Minute-used: {self.minUsed} minutes')
            print(f'Monthly payment: {self.calPayment()} baht')
            print('------------------------------------')
        else:
            print('Cannot make a call, please re-charge your mobile')
            print('------------------------------------')
    def charge(self,mins):
        self.remainingBattery = self.remainingBattery + (100*mins) if self.remainingBattery+(100*mins)<=self.battery else self.battery
        print(f'charging ({mins} mins)....')
        print(f'battery remaining {self.remainingBattery} ({self.remainingBattery/self.battery*100:.1f}%)')
```

```python
# exercises 5 TEST
mb1 = Mobile('Samsung',(1080,1340),5000,(800,1800))
print(mb1)
mb1.call(20)
mb1.call(4000)
mb1.call(20)
mb1.charge(10)
```

```python
# exercises 6
class Vehicle:
    def __init__(self,brand,model,year):
        self.brand = brand
        self.model = model
        self.year = year
    def display_info(self):
        print(f'Brand: {self.brand}')
        print(f'Model: {self.model}')
        print(f'Year: {self.year}')
```

```python
# exercises 7
class Car(Vehicle):
    def __init__(self, brand, model, year, number_of_seats, fuel_efficiency):
        super().__init__(brand, model, year)
        self.number_of_seats = number_of_seats
        self.fuel_efficiency = fuel_efficiency
    def calculate_fuel_cost(self,distance,fuel_price):
        cost = distance / self.fuel_efficiency * fuel_price
        return cost
    def get_fuel_cost(self,distance,fuel_price):
        return self.calculate_fuel_cost(distance,fuel_price)
    def display_info(self):
        super().display_info()
        print(f'Number of seats: {self.number_of_seats}')
```

```python
# exercises 8
class Truck(Vehicle):
    def __init__(self, brand, model, year, cargo_capacity):
        super().__init__(brand, model, year)
        self.cargo_capacity = cargo_capacity
        self.load_status = False
    def check_load(self):
        return self.load_status
    def update_load_status(self,status):
        self.load_status = status
        print(f'Load status: {self.load_status}')
    def display_info(self):
        super().display_info()
        print(f'Cargo capacity: {self.cargo_capacity} kg')
        print(f'Load status: {'Loaded' if self.load_status else 'Not Loaded'}')
```

```python
# exercises 9
class Motorcycle(Vehicle):
    def __init__(self, brand, model, year, engine_type, max_speed):
        super().__init__(brand, model, year)
        self.engine_type = engine_type
        self.max_speed = max_speed
    def chack_speed(self, speed):
        self.is_speed_safe(speed)
    def is_speed_safe(self, speed):
        if speed >= 0:
            if speed == self.max_speed:
                return 'No, the speed is unsafe'
            elif speed > 0.9 * self.max_speed:
                return 'Warning: Speed is near maximum limit. Drive cautiously'
            else:
                return 'Yes, the speed is safe'
        else:
            return 'Invalid speed. Speed must be positive'
    def display_info(self):
        super().display_info()
        print(f'Engine type: {self.engine_type}')
        print(f'Maximum speed: {self.max_speed} km/h')
```

```python
# Example usage
car = Car('Toyota', 'Corolla', 2020, 5, 15)
truck = Truck('Fort', 'F-150', 2019, 3000)
motorcycle = Motorcycle('Honda','CBR500R',2021,'4-Stroke',180)

print('Car Information:')
car.display_info()
print(f'Fuel cost for 100 km at 40 THB/L: {round(car.calculate_fuel_cost(100,40),2)} THB')

print('\nTruck Information:')
truck.display_info()
truck.load_status = True
truck.display_info()

print(f'\nMotorcycle Information:')
motorcycle.display_info()
print(f'Is speed 200 km/h safe? {motorcycle.is_speed_safe(200)}')
print(f'Is speed 150 km/h safe? {motorcycle.is_speed_safe(150)}')
```
